```python
import pandas as pd
import plotly.express as px
```

```python
#Planilha usada para fazer Análise
dfPrincipal = pd.read_excel("Acoes.xlsx", sheet_name = "Principal")
dfTotalAcoes = pd.read_excel("Acoes.xlsx", sheet_name = "Total_de_acoes")
dfTicker = pd.read_excel("Acoes.xlsx", sheet_name="Ticker")
```

## Tratamento de Dados

Antes de realizar as análises propriamente ditas, é necessário fazer um pré-tratamento dos dados, onde eles serão organizados e estruturados de maneira mais eficiente, com as informações necessárias para fazer a análise.

```python
#Exclui as colunas que estão além de VAr
dfPrincipal = dfPrincipal[["Ativo", "Data", "Último (R$)", "Var. Dia (%)"]].copy()

#Renomeia colunas
dfPrincipal = dfPrincipal.rename(columns={'Último (R$)':'Valor_final', 'Var. Dia (%)':'Var_dia_pct'}).copy()
dfPrincipal
```

```python
dfPrincipal["Var_pct"] = dfPrincipal["Var_dia_pct"]/100
dfPrincipal
```

```python
#Transforma o valor em porcentagem para decimal
dfPrincipal["Valor_inicial"] = dfPrincipal["Valor_final"]/(dfPrincipal["Var_pct"]+1)
dfPrincipal
```

```python
#Une as planilhas principal e total de ações por meio das colunas Ativo e Código.
dfPrincipal = dfPrincipal.merge(dfTotalAcoes, left_on='Ativo', right_on='Código', how='left')
dfPrincipal
```

```python
#Exclui a coluna repetida Código
dfPrincipal = dfPrincipal.drop(columns = ["Código"])
dfPrincipal
```

```python
dfPrincipal = dfPrincipal.rename(columns = {"Qtde. Teórica": "Qtd_acoes"}).copy()

#Calcula a variação em reais das ações
dfPrincipal["Variacao_reais"] = (dfPrincipal["Valor_final"]-dfPrincipal["Valor_inicial"])*dfPrincipal["Qtd_acoes"]

dfPrincipal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
dfPrincipal["Qtd_acoes"] = dfPrincipal["Qtd_acoes"].astype(int)
dfPrincipal
```

```python
#Aplica a função lambda a cada linha da planilha, gerando uma nova coluna resultado
dfPrincipal['Resultado'] = dfPrincipal['Variacao_reais'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
dfPrincipal
```

```python
#Pega agora o nome de cada empresa a partir da planilha dfTicker
dfPrincipal = dfPrincipal.merge(dfTicker, left_on='Ativo', right_on='Ticker', how='left')
dfPrincipal = dfPrincipal.drop(columns=['Ticker'])
dfPrincipal
```

```python
dfGPT = pd.read_excel("Acoes.xlsx", sheet_name="Chatgpt")
dfGPT
```

```python
dfPrincipal = dfPrincipal.merge(dfGPT, left_on = "Nome", right_on = "Empresa", how = "left")
dfPrincipal = dfPrincipal.drop(columns = ["Empresa"]).copy()
dfPrincipal
```

```python
dfPrincipal['Cat_idade'] = dfPrincipal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
dfPrincipal
```

```python
# Calculando o maior valor
maior = dfPrincipal['Variacao_reais'].max()

# Calculando o menor valor
menor = dfPrincipal['Variacao_reais'].min()

# Calculando a média
media = dfPrincipal['Variacao_reais'].mean()

# Calculando a média de quem subiu
media_subiu = dfPrincipal[dfPrincipal['Resultado'] == 'Subiu']['Variacao_reais'].mean()

# Calculando a média de quem desceu
media_desceu = dfPrincipal[dfPrincipal['Resultado'] == 'Desceu']['Variacao_reais'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
dfSubiu = dfPrincipal[dfPrincipal["Resultado"]== "Subiu"]
dfSubiu
```

```python
dfSetorSubiu = dfSubiu.groupby("Setor")["Variacao_reais"].sum().reset_index(0)
dfSetorSubiu
```

```python
dfSaldo = dfPrincipal.groupby("Resultado")["Variacao_reais"].sum().reset_index()
dfSaldo
```

```python
dfIdade = dfPrincipal.groupby("Cat_idade")["Variacao_reais"].sum().reset_index()
dfIdade
```

```python
nome_arquivo = 'PlanilhaTratada.xlsx'
writer = pd.ExcelWriter(nome_arquivo, engine='xlsxwriter')

# Salvando cada DataFrame como uma planilha separada no arquivo Excel
dfPrincipal.to_excel(writer, sheet_name='PlanilhaPrincipal', index=False)  
dfSubiu.to_excel(writer, sheet_name='PlanilhaSubiu', index=False)
dfSetorSubiu.to_excel(writer, sheet_name='PlanilhaSetor', index=False)
dfSaldo.to_excel(writer, sheet_name='PlanilhaSalso', index=False)
dfIdade.to_excel(writer, sheet_name='PlanilhaIdade', index=False)
# Adicione mais linhas para salvar mais planilhas, se necessário

# Fechando o escritor do Excel
writer.close()
```
